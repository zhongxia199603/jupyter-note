```python
def _shuffle(a, b, c):
    d = list(zip(a, b, c))
    random.shuffle(d)
    a[:], b[:], c[:] = zip(*d)
    return a, b, c
```

```python
keywords, candidate, reference = _shuffle(keywords, candidate, reference)
```
