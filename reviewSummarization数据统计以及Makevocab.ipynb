### 文件读取

```python
path = '/home/jinhanqi/tf_program/summarization_data/review_data/Toys_and_Games_5_train.json'
```

```python
vocab_counter = collections.Counter()
with open(path, 'r') as f:
    content = f.readlines()
```

```python
reviews = []
summarys = []
```

```python
for line in content:
    dic = json.loads(line)
    reviews.append(dic['reviewText'])
    summarys.append(dic['summary'])
```

### 制作词汇表

```python
i = 0
for article, abstract in zip(reviews, summarys):
    if i % 100000 == 0:
        print i
    i += 1
    art_tokens = article.split(' ')
    abs_tokens = abstract.split(' ')
    #abs_tokens = [t for t in abs_tokens if t not in [SENTENCE_START, SENTENCE_END]] # remove these tags from vocab
    tokens = art_tokens + abs_tokens
    tokens = [t.strip() for t in tokens] # strip
    tokens = [t for t in tokens if t!=""] # remove empty
    vocab_counter.update(tokens)
```

### 将词汇表写入文件

```python
vocab_path = '/home/jinhanqi/tf_program/summarization_data/review_data'
```

```python
print "Writing vocab file..."
with open(os.path.join(vocab_path, "vocab"), 'w') as writer:
    #for word, count in vocab_counter.most_common(VOCAB_SIZE):
    #    writer.write(word + ' ' + str(count) + '\n')
    for word, count in vocab_counter.most_common(len(vocab_counter)):
        try:
            writer.write(word + ' ' + str(count) + '\n')
        except ValueError:
            print word, count
print "Finished writing vocab file"
```
