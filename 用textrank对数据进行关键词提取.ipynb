```python
from summa import keywords
def sortByOrder(text, a, b):
    if text.find(a) == -1:
        _a = a.split(' ')[0]
    else:
        _a = a
    if text.find(b) == -1:
        _b = b.split(' ')[0]
    else:
        _b = b
    ind_a = text.index(_a)
    ind_b = text.index(_b)
    return ind_a < ind_b
def Get_keywords_and_sort_with_ratio(text, words_ratio=1):
    _keyword = keywords.keywords(text, ratio=words_ratio)
    _keyword = _keyword.split('\n')
    _keyword = [w if w!='unk' else 'UNK' for w in _keyword ]   # 测试数据中有UNK
    _keyword = [w for w in _keyword if text.find(w) !=-1 ]  # 去掉未在原文中出现的词
    keyword = sorted(_keyword, lambda x,y: -sortByOrder(text, x, y))
    keyword = ' '.join(keyword)
    keyword = keyword.split(' ')
    keyword = sorted(set(keyword), key = keyword.index)   # 去掉重复元素
    return keyword
```

```python
with open(article_path, 'r') as f:
    articles = f.readlines()
print len(articles)
```

```python
print articles[0]
```

```python
Keywords = []  
for i, t in enumerate(articles):
    key = Get_keywords_and_sort_with_ratio(t, 1.0)
    Keywords.append(key)
    if i % 5000 == 0:
        print i, key
```

```python
with open(output_path, 'w') as f:
    for key in Keywords:
        f.write(' '.join(key)+'\n')
```
